```python
import re
import pandas as pd
import spacy
# Removed: from joblib import Parallel, delayed

# Load spaCy model untuk Bahasa Indonesia
nlp = spacy.load('xx_ent_wiki_sm')  # Bahasa Indonesia model di spaCy

# Load dataset
df = pd.read_csv('scrapping_hok.csv')

def clean_text_spacy(text):
    # Handle NaN/None values
    if pd.isna(text):
        return ""

    # Convert to string if not already
    text = str(text)

    # Hapus URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    # Hapus tag HTML
    text = re.sub(r'<.*?>', '', text)

    # Hapus emoji dan karakter khusus
    text = re.sub(r'[^\w\s]', '', text)

    # Hapus angka
    text = re.sub(r'\d+', '', text)

    # Ubah ke huruf kecil
    text = text.lower()

    # Tokenisasi dan Stopword removal dengan spaCy
    doc = nlp(text)
    tokens = [token.text for token in doc if not token.is_stop and len(token.text) > 2]

    # Gabungkan kembali token menjadi satu string
    text = ' '.join(tokens)

    # Hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Removed: # Fungsi untuk menerapkan clean_text secara paralel
# Removed: def parallel_cleaning(texts):
# Removed:     return Parallel(n_jobs=-1)(delayed(clean_text_spacy)(text) for text in texts)

# Apply the cleaning function directly using pandas' apply
df['cleaned_text'] = df['content'].apply(clean_text_spacy) # Changed to apply to avoid parallel processing

# Hapus baris yang kosong setelah cleaning
df = df[df['cleaned_text'].str.len() > 0]

# Simpan hasil
df.to_csv("cleaned_scrapping_hok.csv", index=False)
```

**WORD2VEC**

```python
from nltk.tokenize import word_tokenize
import numpy as np
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

X_train_tokens = X_train.apply(word_tokenize)
X_test_tokens = X_test.apply(word_tokenize)
```

```python
from gensim.models import Word2Vec

word2vec_model = Word2Vec(
    X_train_tokens,
    vector_size=300,  # Dimensi vektor yang lebih besar
    window=10,        # Konteks yang lebih luas
    min_count=5,      # Buang kata yang terlalu jarang
    sg=1,             # Gunakan Skip-gram
    workers=4         # Paralelisasi
)
```

```python
def get_word2vec_features(tokens, model):
    features = []
    for word in tokens:
        if word in model.wv:
            features.append(model.wv[word])
    if not features:
        return np.zeros(model.vector_size)
    else:
        return features

# Mengonversi teks menjadi vektor untuk data training dan testing
X_train_w2v = X_train_tokens.apply(lambda x: get_word2vec_features(x, word2vec_model))
X_test_w2v = X_test_tokens.apply(lambda x: get_word2vec_features(x, word2vec_model))

# Rata-rata vektor kata untuk setiap review
X_train_w2v_avg = X_train_w2v.apply(lambda x: np.mean(x, axis=0) if len(x) > 0 else np.zeros(word2vec_model.vector_size))
X_test_w2v_avg = X_test_w2v.apply(lambda x: np.mean(x, axis=0) if len(x) > 0 else np.zeros(word2vec_model.vector_size))
```

```python
print("W2V Training :", X_train_w2v.shape)
print("W2V Testing:", X_test_w2v.shape)
print("W2V Training AVG :", X_train_w2v_avg.shape)
print("W2V Testing AVG:", X_test_w2v_avg.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Convert X_train_w2v_avg and X_test_w2v_avg to NumPy arrays with consistent shape
X_train_w2v_avg_np = np.vstack(X_train_w2v_avg.apply(lambda x: x if isinstance(x, np.ndarray) else np.zeros(word2vec_model.vector_size)).values) # Convert to NumPy array, handling non-array elements
X_test_w2v_avg_np = np.vstack(X_test_w2v_avg.apply(lambda x: x if isinstance(x, np.ndarray) else np.zeros(word2vec_model.vector_size)).values) # Convert to NumPy array, handling non-array elements

# Melatih model menggunakan rata-rata fitur Word2Vec
clf_w2v = LogisticRegression(max_iter=1000)
clf_w2v.fit(X_train_w2v_avg_np, y_train)  # Use the NumPy array

# Prediksi dan evaluasi model
y_pred_w2v = clf_w2v.predict(X_test_w2v_avg_np)  # Use the NumPy array
accuracy_w2v = accuracy_score(y_test, y_pred_w2v)
```

```python
print("Akurasi Word2Vec: ", accuracy_w2v)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Prediksi dan evaluasi model
y_pred_w2v = clf_w2v.predict(X_test_w2v_avg_np)
accuracy_w2v = accuracy_score(y_test, y_pred_w2v)
print("Akurasi Word2Vec: ", accuracy_w2v)

# Print classification report
print(classification_report(y_test, y_pred_w2v))

# Create and display the confusion matrix
cm = confusion_matrix(y_test, y_pred_w2v)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['negative', 'neutral', 'positive'],
            yticklabels=['negative', 'neutral', 'positive'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
